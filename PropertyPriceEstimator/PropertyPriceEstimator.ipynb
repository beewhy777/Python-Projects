```python
from tkinter import *
from tkinter import ttk
import pandas as pd
win =Tk()
win .title("PROPERTY PRICE ESTIMATOR") 
photo = PhotoImage(file = "Dynamo_Logo.png")
win.iconphoto(False, photo)
win.geometry = ("400,350")
win.resizable(False,False)
win.config(bg="#d6d6cb")

# Function to estimate price of property
def estimate():
    try:
        prop = pd.read_excel("myprop.xlsx")
        prop.set_index(["AreaCode","Type"],inplace =True)
        AreaCode = postal_entry.get().replace(" ","") .upper()
        Type = building_type.get().title()
        price = prop.loc[(AreaCode,Type),"Price"]
        no_of_beds = int(bed.get())
        no_of_garages = int(garage.get())
        no_of_bathrooms = int(bathroom.get())
        jacuzzi = cb1.get()
        pool = cb2.get()
        basement = cb3.get()
        price = price + (no_of_beds * 5000) 
        price = price +  (no_of_garages * 2000) 
        price = price +  (no_of_bathrooms * 5000) 
        if pool == 1:
            price += 10000
        if jacuzzi == 1 :
            price += 10000 
        if basement == 1 :
            price += 15000 
        home_insurance = price * 0.10
        price = price + home_insurance
        answer = f"The Price for Property in the area \nyou selected is ${price:,.2f}"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid entry entered or you \nhave entered an unavailable Postal Code"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
        
# Function to Reinitialize the App        
def zerorize():
    cb1.set(0)
    cb2.set(0)
    cb3.set(0)
    answerlabel.config(text="")
    postal_entry.delete(0,END)
    bed.set("")
    garage.set("")
    bathroom.set("")
    building_type.set("")
    
# Creation of the App's Frames     
frame1 =Frame(win,width=400,height=50,bg="white")
frame1.pack(side=TOP)
frame2 =Frame(win,width=400,height=50,bg="#d6d6cb")
frame2.pack(side=TOP)
frame3 =Frame(win,width=400,height=50,bg="#d6d6cb")
frame3.pack(side=TOP)
frame4 =Frame(win,width=400,height=50,bg="#d6d6cb")
frame4.pack(side=TOP)
frame5 =Frame(win,width=400,height=50,bg="#d6d6cb")
frame5.pack(side=TOP)
frame6 =Frame(win,width=400,height=50,bg="#d6d6cb")
frame6.pack(side=TOP)
frame7 =Frame(win,width=400,height=50,bg="#d6d6cb")
frame7.pack(side=TOP)
frame8 =Frame(win,width=400,height=50,bg="#d6d6cb")
frame8.pack(side=TOP)

# Creation of the App's Widgets
answerlabel = Label(frame1,bg="white",fg="black",width=40,height=3,bd=2,relief=RAISED)
answerlabel.grid(row=0,column=0)


postal_label = Label(frame2,text = "Postal Code",width =16,height=1,font=("Teveltica",16)
                     ,fg="black",bg="#ededdf",anchor="w",justify="left")
postal_label.grid(row=0,column=0,padx=8,pady=5)

var1 = StringVar()
postal_entry = Entry(frame2,textvariable=var1,font=("Teveltica",14),fg="black",width=12,bg="white")
postal_entry.grid(row=0,column=1,padx=4,pady=5,rowspan=2)

type_label = Label(frame3,text = "Building Type",width =16,height=1,font=("Teveltica",16),
                   fg="black",bg="#ededdf",anchor="w",justify="left")
type_label.grid(row=0,column=0,padx=8,pady=5)

var2= StringVar()
building_type = ttk.Combobox(frame3, width = 10, textvariable = var2,font=("Teveltica",14))
building_type .config(height=500)
building_type .grid( row = 0,column = 1,padx=4,pady=5)
building_type .current()
building_type['values'] = ("Bungalow", "Duplex", "Triplex")

bed_label = Label(frame4,text = "No of Beds",width =9,height=1,font=("Teveltica",14),
                  fg="black",bg="#ededdf")
bed_label .grid(row=0,column=0,pady=5)

mylist=list(range(1,11))
bed = ttk.Combobox(frame4, width = 2,font=("Teveltica",15),values=mylist)
bed .config(height=500)
bed .grid( row = 0,column = 1,padx=5,pady=5)

garage_label = Label(frame4,text = "No of Garages",width =12,height=1,font=("Teveltica",14),
                     fg="black",bg="#ededdf")
garage_label .grid(row=0,column=2,pady=5)

garage = ttk.Combobox(frame4, width =2,font=("Teveltica",15),values=mylist)
garage .config(height=500)
garage .grid( row = 0,column = 3,padx=5,pady=5)


bathroom = ttk.Combobox(frame5, width =2,font=("Teveltica",15),values=mylist)
bathroom.config(height=500)
bathroom .grid( row = 0,column = 2,padx=5,pady=5)


bathroom_label = Label(frame5,text = "No of Bathrooms",width =27,height=1,font=("Teveltica",14),
                     fg="black",bg="#ededdf")
bathroom_label .grid(row=0,column=0,columnspan=2,pady=5)

facility_label=Label(frame6,text ="Optional Facilities",width=25,height=1,bg="#ededdf",
                font =("Teveltica",14),bd=2,relief=RAISED)
facility_label.grid(row=0,column=0,columnspan=3,padx=15,pady=5)

cb1 = IntVar()  
cb2 = IntVar()  
cb3 = IntVar()
  
jacuzzi = Checkbutton(frame7, text = "Jacuzzi",variable = cb1,onvalue = 1,
          offvalue = 0,height = 2,width = 10,bg="#ededdf",font =("Teveltica",10))
jacuzzi.grid(row=0,column=0,padx=5,pady=2)

swimming_pool = Checkbutton(frame7, text = "Swimming Pool",variable = cb2,onvalue = 1,
          offvalue = 0,height = 2,width = 12,bg="#ededdf",font =("Teveltica",10))
swimming_pool.grid(row=0,column=1,padx=5,pady=2)

basement = Checkbutton(frame7, text = "Basement",variable = cb3,onvalue = 1,
          offvalue = 0,height = 2,width = 10,bg="#ededdf",font =("Teveltica",10))
basement.grid(row=0,column=2,padx=5,pady=2)

button_estimate=Button(frame8, text="ESTIMATE",width=12,command=estimate,
                bg="#ededdf",fg="green",font =("Teveltica",13),relief=RAISED,bd=2)
button_estimate.config(activebackground="#ededdf",activeforeground="green")
button_estimate.grid(row=0,column=0,padx=5,pady=10)

button_zerorize=Button(frame8, text="ZERORIZE",width=12,command=zerorize,
               bg="#ededdf",fg="red",font =("Teveltica",13),relief=RAISED,bd=2)
button_zerorize.config(activebackground="#ededdf",activeforeground="red")
button_zerorize.grid(row=0,column=1,padx=5,pady=10)

win.mainloop()
```
