```python
from tkinter import *
from tkinter import ttk
import pandas as pd
df = pd.read_excel(r"unitconverter.xlsx")
df = df.set_index(["Conversion_Type","From","To"])
win =Tk()
win .title("DC UNIT CONVERTER V2.0") 
photo = PhotoImage(file = "Dynamo_Logo.png")
win.iconphoto(False, photo)
win.geometry = ("400,600")
win.resizable(False,False)
win.config(bg="#AABDCC")

# Bind the measurement to the units' dropdown boxes 
def select_unit(e):
    if measures.get()=="Length":
        From.config(value = mylength)
        To.config(value=mylength)
    elif measures.get()== "Time":
        From.config(value = mytime)
        To.config(value=mytime)
    elif measures.get()== "Mass":
        From.config(value = mymass)
        To.config(value=mymass)
    elif measures.get()== "Volume":
        From.config(value = myvolume)
        To.config(value=myvolume)
    elif measures.get()== "Speed":
        From.config(value = myspeed)
        To.config(value=myspeed)
    elif measures.get()== "Area":
        From.config(value = myarea)
        To.config(value=myarea)
    elif measures.get()== "Digital Storage":
        From.config(value = mydigital)
        To.config(value=mydigital)
    elif measures.get()== "Fuel Consumption":
        From.config(value = myconsumption)
        To.config(value= myconsumption)

# Function to reset all widgets' values and settings
def clear():
    varz.set("")
    var1.set("")
    varx.set("")
    To.set("")
    From.set("")
    entryqty.delete(0, END) 
    answerlabel.config(text="")
    
# Function to convert from one unit to another
def convert():
    try: 
        float(entryqty.get())
        quantity = float(entryqty.get())
        measurement= measures.get()
        from_unit = From.get()
        to_unit = To.get()
        if  measurement == "Mass" :
            mass(measurement,from_unit,to_unit,quantity)
        elif measurement == "Area" :
            area(measurement,from_unit,to_unit,quantity)
        elif measurement =="Time":
            time(measurement,from_unit,to_unit,quantity)
        elif measurement == "Length":
            length(measurement,from_unit,to_unit,quantity)
        elif measurement == "Speed":
            speed(measurement,from_unit,to_unit,quantity)
        elif measurement == "Volume":
            volume(measurement,from_unit,to_unit,quantity)
        elif measurement == "Digital Storage":
            digital(measurement,from_unit,to_unit,quantity)
        elif measurement == "Fuel Consumption":
            consumption(measurement,from_unit,to_unit,quantity)
    except ValueError:
        answer = "The quantity entered not valid.\nEnter a numerical value!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
        
# Functions to calculate and display the results of conversion
def mass(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "A mass of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
        
def area(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "An area of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
def time(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "A time of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    
def length(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "A length of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
def speed(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "An area of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
def volume(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "A volume of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    
def digital(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "A digital storage of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
def consumption(measurement,from_unit,to_unit,quantity):
    try:
        multiplier = df.loc[(measurement,from_unit,to_unit),"Multiply_By"]
        answer = "A fuel consumption of " + str(entryqty.get()) + " "+ From.get() + "\nis equal to "+ "\n"+ str(round(float(multiplier * quantity),4)) +" "+ To.get()
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid or incomplete values entered!!!"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    
# Creation of Frames
frame1 =Frame(win,width=20,height=100,bg="white")
frame1.pack(side=TOP)
answerlabel = Label(frame1,bg="white",fg="black",width=35,height=3,bd=2,relief=RAISED)
answerlabel.grid(row=0,column=0)
frame2 =Frame(win,width=400,height=100,bg="#AABDCC")
frame2.pack(side=TOP)
frame3 =Frame(win,width=400,height=100,bg="#AABDCC")
frame3.pack(side=TOP)
frame4 =Frame(win,width=400,height=100,bg="#AABDCC")
frame4.pack(side=TOP)
frame5 =Frame(win,width=400,height=100,bg="#AABDCC")
frame5.pack(side=TOP)
frame6 =Frame(win,width=400,height=100,bg="#AABDCC")
frame6.pack(side=TOP)

# Creation of other Widgets
labelmeasure=Label(frame2,text ="Measure",width=15,height=1,bd=1,relief=RAISED,bg="#AABDCC")
labelmeasure.config(font=("Arial,1"))
labelmeasure.grid(row=0,column=0,pady=10,padx=5)

var1= StringVar()
measures = ttk.Combobox(frame2, width = 10, textvariable = var1,font=("Arial,1"))
measures.config(height=500)
measures.grid( row = 0,column = 1,pady=10,padx=5)
measures.current()
measures.bind("<<ComboboxSelected>>",select_unit)
measures ['values'] = ("Time", "Mass", "Length", "Volume", "Speed","Area",
                       "Fuel Consumption","Digital Storage")
measures.config(state="readonly") # Prevent Manual Entry of values

entrylabel=Label(frame3,text ="Qty to Convert",width=15,height=1,bg="#AABDCC",bd=2,relief=RAISED)
entrylabel.config(font=("Arial,1"))
entrylabel.grid(row=0,column=0,pady=10,padx=7)

entryqty = Entry(frame3,font=("Arial,bold,10"),width=11,bg="white") 
entryqty.grid(row=0,column=1,pady=10,padx=7)

label_from=Label(frame4,text ="Convert From",width=15,height=1,bg="#AABDCC",bd=2,relief=RAISED)
label_from.config(font=("Arial,1"))
label_from.grid(row=0,column=0,pady=10,padx=5)

varx= StringVar()
From = ttk.Combobox(frame4, width = 10,height=5, textvariable = varx,font=("Arial,1"))
From.grid( row = 0,column = 1,pady=10,padx=5)
From.config(state="readonly") # Prevent Manual Entry of values

label_to=Label(frame5,text ="Convert To",width=15,height=1,bg="#AABDCC",bd=2,relief=RAISED)
label_to.grid(row=0,column=0,pady=10,padx=5)
label_to.config(font=("Arial,1"))

varz= StringVar()
To = ttk.Combobox(frame5, width = 10, textvariable = varz,font=("Arial,1"))
To.grid( row = 0,column = 1,pady=10,padx=5)
To.config(state="readonly") # Prevent Manual Entry of values

button_convert=Button(frame6, text="CONVERT",width=10,command=convert,
bg="#C8D8E6",fg="green",font =("Arial,25,bold"),relief=RAISED,bd=2)
button_convert.config(activebackground="#C8D8E6",activeforeground="green")
button_convert.grid(row=0,column=0,padx=5,pady=10)

button_clear=Button(frame6, text="CLEAR",width=10,command=clear,
                      bg="#C8D8E6",fg= "red",font =("Arial,25,bold"),relief=RAISED,bd=2)
button_clear.config(activebackground="#C8D8E6",activeforeground="red")
button_clear.grid(row=0,column=1,padx=5,pady=10)

# The various categories of Units
mylength =["Kilometer","Meter","Centimeter","Millimeter",
"Mile","Yard","Foot","Inch","Nautical Mile"]
mytime = ["Nanosecond","Microsecond","Millisecond","Second","Minute",
"Hour","Day","Week","Month","Year","Decade","Century"]
mymass =["Metric Ton","Kilogram","Gram","Milligram","Mcg","Long Ton","Short Ton",
"Stone","Pound","Ounce"]
myvolume = ["Us Gal","Us Quart","Us Pint","Us Cup","Us Oz","Us Tbsp.","Us Tsp.",
"Cubic Meter","Liter","Milliliter","Imperial Gal","Imperial Quart","Imperial Pint",
"Imperial Oz","Imperial Tbsp.","Imperial Tsp.","Cubic Foot","Cubic Inch"]
myspeed = ["Miles/Hour","Feet/Sec","Meters/Sec","Km/Hour","Knot"]

myarea =["Square Km","Hectare","Square Meter","Square Mile","Acre","Square Yard","Square Foot",
"Square Inch"]

mydigital=["Bit","Byte","Kilobit","Kilobyte","Kibibit","Kibibyte","Megabit","Megabyte",
"Mebibit","Mebibyte","Gigabit","Gigabyte","Gibibit","Gibibyte","Terabit",
"Terabyte","Tebibit","Tebibyte","Petabit","Petabyte","Pebibit","Pebibyte"]
myconsumption =["Mpg(Us)","Mpg(Imp.)","Km/Liter"]

win.mainloop()









```
